```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline as uspl
from scipy import optimize as opt 
import scipy
import scipy.odr as odr
from scipy import interpolate
import plotly.plotly as py
import plotly.graph_objs as go
from matplotlib import rc

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

data1 = pd.read_csv('experiment1CdSe.csv')
data2 = pd.read_csv('experiment2CdS.csv')
```

```python
x1_in=data1['gradphi']
y1_in=data1['u']
x2_in=data2['gradphi']
y2_in=data2['u']
```

```python
data1
```

```python
data2
```

```python
#CdSe
x1 = np.array(x1_in)
y1 = np.array(y1_in)
f = interpolate.interp1d(x1, y1, kind='quadratic')
#CdS
x2 = np.array(x2_in)
y2 = np.array(y2_in)
g = interpolate.interp1d(x2, y2, kind='quadratic')
```

```python
xnew1 = np.arange(104.1+570, 154.8+570, 0.01)
ynew1 = f(xnew1)   # use interpolation function returned by `interp1d`
plt.plot(x1, y1, 'o', xnew1, ynew1, '-')

plt.xlabel(r"$\lambda$,nm",fontsize=16)
plt.ylabel(r'mV',fontsize=16)
plt.title("Spectral dependancy of photocurrent for CdSe",
    fontsize=16, color='black')
plt.savefig('CdSe',dpi=500)
plt.show()




xnew2 = np.arange(103.88+570, 500+570, 0.01)
ynew2 = g(xnew2)   # use interpolation function returned by `interp1d`
plt.plot(x2[:-1], y2[:-1], 'o', xnew2, ynew2, '-')
plt.xlabel(r"$\lambda$,nm",fontsize=16)
plt.ylabel(r'mV',fontsize=16)
plt.title("Spectral dependancy of photocurrent for CdS",
    fontsize=16, color='black')
plt.savefig('CdS',dpi=500)
plt.show()


plt.plot(x1, y1, 'o', xnew1, ynew1, '-')
plt.plot(x2[:-1], y2[:-1], 'o', xnew2, ynew2, '-')
plt.xlabel(r"$\lambda$,nm",fontsize=16)
plt.ylabel(r'mV',fontsize=16)
plt.title("Spectral dependancy of photocurrent for CdSe+CdS",
    fontsize=16, color='black')
plt.savefig('CdSe+CdS',dpi=500)
plt.show()
```
