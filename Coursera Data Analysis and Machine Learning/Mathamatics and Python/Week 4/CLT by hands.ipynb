```python
from scipy.stats import chi2
from statistics import mean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
x = np.linspace(0,100,1000)
df=50
rv = sts.chi2(df)
pdf = rv.pdf(x)
r=sts.chi2.rvs(df,size=1000)
plt.plot(x, pdf, label="$k=%s$" % df)
plt.hist(r, density=True, histtype='stepfilled',facecolor='#FF6103',bins=20 ,alpha=0.5)
plt.ylabel("$f(x)$")
plt.xlabel("x")
plt.legend()
plt.title("Функция плотности вероятности распределения + гистограмма  ($\chi^2_k$)")
```

```python
# Cравнение между собой гистограмм для разных n

numbers=[5,10,50,100]
for n in numbers: 
    hist=[mean(sts.chi2.rvs(df,size=n)) for i in range(1000)]
    plt.hist(hist,density=True, histtype='stepfilled',bins=15 ,alpha=0.6, label="$n=%s$" % n)
    plt.legend()
plt.title("Гистограмма распределения выборочного среднего для $\chi^2_k$")
plt.xlabel("Значения выборки")
plt.ylabel("Доля от всех")
plt.show()   
```

Для $\chi^2_k$ матожидание равно $\mu=k$ а дисперсия равна $\sigma^2=2k$ 
$\rightarrow se=\frac{\sigma}{\sqrt{n}}$

```python
# генерация графиков с гистокграммами и аппроксимирущим распределением N(k,se)

numbers=[5,10,50,100]
for n in numbers: 
    hist=[mean(sts.chi2.rvs(df,size=n)) for i in range(1000)]
    plt.hist(hist,density=True, histtype='stepfilled',bins=15 ,facecolor='#76EE00',alpha=0.6,label="$n=%s$" % n)
    norm_rv = sts.norm(loc=df, scale=(2*df/n)**0.5)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
    plt.legend()
    plt.xlabel("Значения выборочного среднего")
    plt.ylabel("Доля от всех")
    plt.title("Распределение выборочного среднего для $\chi^2_k$")    
    plt.show()   
```

Как легко видеть из графиков с ростом $n$ точность аппроксимации растёт из чего видно как работает ЦПТ
